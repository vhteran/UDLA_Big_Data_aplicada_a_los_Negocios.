## Introducci칩n
En esta lecci칩n revisaremos los **sistemas de recomendaci칩n** o tambi칠n conocidos en ingl칠s como _recommender systems_. Los **sistemas de recomendaci칩n** son algoritmos que permiten a las empresas detectar las necesidades y preferencias de sus clientes de forma personalizada, a trav칠s del an치lisis de grandes cantidades de datos. ### Modelos de recomendaci칩n 

El principal objetivo de estos modelos es descubrir patrones en el conjunto de datos donde el sistema recomienda 칤tems al usuario bas치ndose exclusivamente en su experiencia pasada y su la correlaci칩n con sus necesidades e intereses. #### Tipos de modelos de recomendaci칩n

**1. Popularity (basado en popularidad)**: Es un tipo de sistema de recomendaci칩n que funciona seg칰n el principio de popularidad. Estos sistemas comprueban qu칠 productos o pel칤culas est치n de moda o son los m치s populares entre los usuarios y los recomiendan directamente. **2. Content-based filtering (filtrado basado en el contenido)**: Es otro tipo de sistema de recomendaci칩n que funciona seg칰n el principio del contenido similar. Si un usuario est치 viendo una pel칤cula, el sistema buscar치 otras pel칤culas de contenido similar o del mismo g칠nero que la pel칤cula que est치 viendo. Ejemplo: Google que a partir del hist칩rico de b칰squedas o visitas realizadas nos recomienda web similares.
 
**3. Collaborative filtering (filtrado colaborativo)**:  Est치 considerado como uno de los sistemas de recomendaci칩n m치s inteligentes que funcionan a partir de la similitud entre diferentes usuarios y tambi칠n art칤culos que se utilizan ampliamente como sitio web de comercio electr칩nico y tambi칠n sitios web de pel칤culas en l칤nea. Comprueba los gustos de usuarios similares y hace recomendaciones. La similitud no se limita al gusto del usuario, sino que tambi칠n se puede tener en cuenta la similitud entre diferentes art칤culos. El sistema ofrecer치 recomendaciones m치s eficaces si disponemos de un gran volumen de informaci칩n sobre usuarios y art칤culos. Ejemplo: Amazon cuyo modelo de recomendaci칩n identifica usuarios similares y recomienda nuevos 칤tems a otros usuarios basado en el rating dado por otros usuarios similares (que no haya valorado este usuario).

# 5. Validaci칩n y limpieza de datos
Los datos no v치lidos pueden sesgar y falsear los resultados de los an치lisis. La etapa de validaci칩n y depuraci칩n de datos se enfoca en establecer reglas de validaci칩n a menudo complejas y a eliminar cualquier dato no v치lido conocido.

```python
#Seleccionamos las columnas que requerimos para el an치lisis y renombra con: 'customerId', 'purchase_count', 'productId'
dataset = data[["transaction","quantity","item"]]
display(dataset)
```

```python
# Renombramos los columnas seleccionadas previamente por 'customerId', 'purchase_count' y 'productId'
dataset.columns=['customerId', 'purchase_count', 'productId']
display(dataset)
```

```python
# Agrupamos los datos por 'customerId' y 'productId'
dataset_transformado = dataset.groupby(['customerId', 'productId'])['purchase_count'].sum()
dataset_transformado.head()
```

```python
# Transponemos el dataset
dataset_transformado = dataset_transformado.unstack()
dataset_transformado.head()
```

**Normalizar los valores de los elementos entre los usuarios**

Para hacer esto, normalizamos la frecuencia de compra de cada art칤culo entre los usuarios creando primero una matriz usuario-art칤culo de la siguiente manera:

游눠 **Normalizar:** significa que se van a ajustar los valores medidos en diferentes escalas respecto a una escala com칰n. Ayudar치 a mejorar el performance del modelo._

```python
#Creamos una matriz normalizada
df_matrix_norm = (dataset_transformado-dataset_transformado.min())/(dataset_transformado.max()-dataset_transformado.min())
df_matrix_norm.head()
```

**Crear una tabla para la entrada al modelado**

```python
#Agregamos como columna el campo 'customerId'
dataModelo = df_matrix_norm.reset_index() 
dataModelo.head()
```

```python
#Cambiamos el nombre de la columna index por 'scaled_purchase_freq'
dataModelo.index.names = ['scaled_purchase_freq'] 
dataModelo.head()
```

```python
#Agregamos como columna el campo 'customerId'
dataModelo = df_matrix_norm.reset_index() 
dataModelo.head()
```

```python
#Obtenemos una matriz usuario, producto, rating/interacci칩n
dataModelo = pd.melt(dataModelo, id_vars=['customerId'], value_name='scaled_purchase_freq').dropna()
dataModelo.head()
```

```python
dataModelo.scaled_purchase_freq.mean()
```

**Definir modelos usando la librer칤a Turicreate**

```python
user_id = 'customerId'
item_id = 'productId'
users_to_recommend = list(dataset[user_id])

n_rec = 10 #n칰mero de items recomendados
n_display = 20 #visualizar las primeras 20 filas del resultado
```

## Implementaci칩n modelo Popularity (basado en popularidad)
Como revisamos anteriormente, el modelo **Popularity** es la manera m치s sencilla de crear un sistema de recomendaci칩n. Entonces, 쯖칩mo identificamos los productos populares? 쮺u치les son los productos que tienen mejor puntaje/rating?

```python
#Crear modelo por popularidad
popularity = tc.popularity_recommender.create(tc.SFrame(dataModelo), 
                                                    user_id = user_id, 
                                                    item_id = item_id, 
                                                     target='scaled_purchase_freq')
#Recomendaci칩n de productos por popularidad
recom = popularity.recommend(users=users_to_recommend, k=n_rec)
recom.print_rows(n_display)
```
